```python
## json Test
import pandas as pd
import json

train_data = pd.read_csv("../../dataset/train/train.csv")
```

```python
def json_text(train_data):
    
    sub_word = []
    sub_start = []
    sub_end = []
    sub_type = []
    
    obj_word = []
    obj_start = []
    obj_end = []
    obj_type = []
    
    for i,j in zip(train_data["subject_entity"], train_data["object_entity"]):
        
        s = i[1:-1].split(":")
        o = j[1:-1].split(":")
        
        s_word = s[1][2:-14]
        s_start = s[2][1:-11]
        s_end = s[3][1:-8]
        s_type = s[4][2:-1]
        
        o_word = o[1][2:-14]
        o_start = o[2][1:-11]
        o_end = o[3][1:-8]
        o_type = o[4][2:-1]
        
        sub_word.append(s_word)
        sub_start.append(s_start)
        sub_end.append(s_end)
        sub_type.append(s_type)
        
        obj_word.append(o_word)
        obj_start.append(o_start)
        obj_end.append(o_end)
        obj_type.append(o_type)
        
    train_data["sub_word"] = sub_word
    train_data["sub_start"] = sub_start
    train_data["sub_end"] = sub_end
    train_data["sub_type"] = sub_type
    
    train_data["obj_word"] = obj_word
    train_data["obj_start"] = obj_start
    train_data["obj_end"] = obj_end
    train_data["obj_type"] = obj_type
    
json_text(train_data)
train_data
```

```python
store = []
obj = list(train_data["obj_word"])
sub = list(train_data["sub_word"])
sentence = list(train_data["sentence"])
for i in range(len(train_data)):
    store.append(obj[i]+" [SEP] "+sub[i]+" [SEP] "+sentence[i])
train_data["sentence"] = store
train_data["sentence"][100]
```
